```python
import pandas as pd
import os

wd = "../data/"

# read datasets from jbloom lab
files = ["human_sera_raw_data.txt",
         "Ellebedy_invivo_dms-view_data.txt",
         "MAP_paper_antibodies_raw_data.txt",
         "REGN_and_LY-CoV016_raw_data.txt",
         "AZ_cocktail_raw_data.txt",
         "Ellebedy_invivo_raw_data.txt"]

datasets = []
for i in files:
    d = pd.read_csv(os.path.join(wd,i))
    d["study"] = [i]*len(d)
    datasets.append(d)

d = pd.concat(datasets,ignore_index=True)
```

```python
#use only human plasma data from human_sera_raw_data.txt
hum = d.query ("study=='human_sera_raw_data.txt'")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
def make_plot(recurent_escapers,hum_rnd_ctrl,subplot):
    a,b,c = subplot
    plt.subplot(a,b,c)
    plot_data = pd.concat([recurent_escapers,hum_rnd_ctrl])
    plot_data["type"] = ["observed"]*len(recurent_escapers) + ["ctrl"]*len(hum_rnd_ctrl)
    sns.histplot(data=plot_data,x="N_samples",hue="type")
```

```python
import numpy as np
q = 95 # percentile of the mut_escape score
percentiles = [50,75,90,95,99]
for n,q in enumerate(percentiles):
    grupped = hum.groupby("condition")
    escapers = grupped.mut_escape.transform(lambda x: x>np.percentile(x,q)).rename("escapers")
    hum_escapers = pd.concat ([hum, escapers], axis=1).query("escapers==True")

    mut_bysite = hum_escapers.groupby(["wildtype","site","mutation"])
    recurent_escapers = mut_bysite.agg("size").reset_index().rename(columns={0:"N_samples"}).sort_values(by="N_samples",ascending=False)

    control = escapers.sample(frac=1).reset_index(drop=True)
    hum_rnd_ctrl = pd.concat ([hum, control], axis=1).query("escapers==True")
    hum_rnd_ctrl = hum_rnd_ctrl.groupby(["wildtype","site","mutation"]).agg("size").reset_index().rename(columns={0:"N_samples"}).sort_values(by="N_samples",ascending=False)
    make_plot(recurent_escapers,hum_rnd_ctrl,(1,1,1))
    
    recurent_escapers.to_csv("results/"+"recurent_escapres_from_human_sera_raw_data.txt_q_"+str(q)+".csv",index=False)
    
    plt.title("q="+str(q)+" N_escapers="+str(mut_bysite.ngroups))
    plt.savefig("results/"+str(q),dpi=300) 
    plt.clf()
```
